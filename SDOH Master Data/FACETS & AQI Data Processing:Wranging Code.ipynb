```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv("/Users/wwaynerichards/Desktop/Biomedical Informatics Degree/Data Wrangling/Group Project/SDOH Master Data/MASTER_FACETS_DATA_AllColumns.csv")

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha = "center")

new_data = pd.DataFrame(columns = ["Column_Names","Count"])
for column_name in data.columns:
    column = data[column_name]
    count = (column == 0).sum()
    new_data = new_data.append({"Column_Names":column_name, "Count":count}, ignore_index = True)

new_data = pd.Series(new_data["Count"].values, index = new_data["Column_Names"])
new_data = new_data.to_dict()
    
plt.style.use('ggplot')
variables = list(new_data.keys())
variable_values = list(new_data.values())
fig = plt.figure(figsize = (20, 8))
plt.xticks(rotation = 90)
plt.bar(variables, variable_values)
#plt.ylim(-1,850)
plt.xlabel("Variable Names", size = 14)
plt.ylabel("No. of Missing Values", size = 14)
plt.title("SDOH Variables - 'Zero' Value Counts",size = 20, fontweight='bold')
addlabels(variables, variable_values, )
plt.show()
```
